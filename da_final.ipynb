##2.2 Data cleanning

### 2.2.1 Listing data import & cleanning

```python
#import data
df = pd.read_csv('listings.csv')

#get rid of the listings with no review in the last twelve months
df = df[df["number_of_reviews_ltm"] > 0]

#convert listing_id to string
df['id'] = df['id'].astype(int)
df['id'] = df['id'].astype(str)

#convert host_response_rate into percentage
df['host_response_rate_%'] = (pd.to_numeric(df['host_response_rate'].str[:-1]).mask(df['host_response_rate'] == '%', 0))

#convert price into dollar
df['price_$'] = df['price'].fillna(0.0).str.replace('[$,]', '').astype('float')

#get rid of the listing with minimum_nights > 365
df = df[df["minimum_nights"] < 365]

#convert instant_bookable into binary (set FALSE as base)
df['instant_bookable_t']=np.where(df['instant_bookable']=='t',1,0)

#convert host_is_superhost into binary (set FALSE as base)
df['host_is_superhost_t']=np.where(df['host_is_superhost']=='t',1,0)

#convert neighbourhood_group_cleansed into binary (set Bronx as base)
df['neighbourhood_group_Manhattan']=np.where(df['neighbourhood_group_cleansed']=='Manhattan',1,0)
df['neighbourhood_group_Brooklyn']=np.where(df['neighbourhood_group_cleansed']=='Brooklyn',1,0)
df['neighbourhood_group_Queens']=np.where(df['neighbourhood_group_cleansed']=='Queens',1,0)
df['neighbourhood_group_SatenIsland']=np.where(df['neighbourhood_group_cleansed']=='Staten Island',1,0)

#convert room_type into binary (set Shared Room as base)
df['room_type_Entire']=np.where(df['room_type']=='Entire home/apt',1,0)
df['room_type_Private']=np.where(df['room_type']=='Private room',1,0)
df['room_type_Hotel']=np.where(df['room_type']=='Hotel room',1,0)

#get rid of the meaningless columns
df = df[['id','name','host_response_rate_%','accommodates','minimum_nights','availability_365',
         'number_of_reviews_ltm','review_scores_rating','instant_bookable_t','host_is_superhost_t',
         'neighbourhood_cleansed','neighbourhood_group_cleansed',
         'neighbourhood_group_Manhattan','neighbourhood_group_Brooklyn','neighbourhood_group_Queens','neighbourhood_group_SatenIsland',
         'room_type','room_type_Entire','room_type_Private','room_type_Hotel',
         'price_$','longitude','latitude']]

#df.head()
```

#4 Sentiment Analysis

```python
# Functions from lecture notes for sentiment analysis
def get_nrc_data():
    nrc = "NRC-emotion-lexicon-wordlevel-alphabetized-v0.92.txt"
    count=0
    emotion_dict=dict()
    with open(nrc,'r') as f:
        all_lines = list()
        for line in f:
            if count < 46:
                count+=1
                continue
            line = line.strip().split('\t')
            if int(line[2]) == 1:
                if emotion_dict.get(line[0]):
                    emotion_dict[line[0]].append(line[1])
                else:
                    emotion_dict[line[0]] = [line[1]]
    return emotion_dict
  
def emotion_analyzer(text,emotion_dict=get_nrc_data()):
    emotions = {x for y in emotion_dict.values() for x in y}
    emotion_count = dict()
    for emotion in emotions:
        emotion_count[emotion] = 0
    total_words = len(text.split())
    for word in text.split():
        if emotion_dict.get(word):
            for emotion in emotion_dict.get(word):
                emotion_count[emotion] += 1/len(text.split())
    return emotion_count

def clean(text):
    lower = text.lower()
    table = str.maketrans("", "", string.punctuation)
    text_cleaned = lower.translate(table)
    return text_cleaned
```

```python
#generate emotion dict from the txt file
emotion_dict = get_nrc_data()
#clean each comment
review['comments'] = review['comments'].apply(lambda x:clean(x)) 
#apply emotion_analyzer to each comment
review['sent'] = review['comments'].apply(lambda x: emotion_analyzer(x))
```

```python
#explode dictionary keys to columns
review = pd.concat([review.drop(['sent'], axis=1), review['sent'].apply(pd.Series)], axis=1)
review
```

```python
review['pos_all']= review['positive']+review['trust']+review['surprise']+review['joy']+review['anticipation']
review['neg_all']= review['fear']+review['anger']+review['negative']+review['sadness']+review['disgust']
review.head()
```

```python
# A comment is good if pos_all is larger than neg_all
review['if_pos'] = np.where(review['pos_all'] > review['neg_all'], 1, 0)
review
```

```python
# Sum the number of comments that are more positive for each listing
# Get the rate of good comments for each listing
gd_rate = review.groupby('listing_id').sum().if_pos / review.groupby('listing_id').size()
gd_rate = pd.DataFrame(gd_rate, columns = ['gd_review_rate'])
gd_rate.reset_index()
```

```python
#Select the meaningful columns
df_ml = new_df[['id','host_response_rate_%','accommodates','minimum_nights','availability_365',
         'numbers_of_reviews_ltm','gd_review_rate','instant_bookable_t','host_is_superhost_t',
         'neighbourhood_group_Manhattan','neighbourhood_group_Brooklyn','neighbourhood_group_Queens','neighbourhood_group_SatenIsland',
         'room_type_Entire','room_type_Private','room_type_Hotel',
         'price_$']]

#Finding the mean of the column having NaN
mean_value=df_ml['host_response_rate_%'].mean()
  
#Replace NaNs in column S2 with the mean of values in the same column
df_ml['host_response_rate_%'].fillna(value=mean_value, inplace=True)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import DataFrame
%matplotlib inline
from sklearn.model_selection import train_test_split

feature_names = ['host_response_rate_%','accommodates','minimum_nights','availability_365',
         'numbers_of_reviews_ltm','gd_review_rate','instant_bookable_t','host_is_superhost_t',
         'neighbourhood_group_Manhattan','neighbourhood_group_Brooklyn','neighbourhood_group_Queens','neighbourhood_group_SatenIsland',
         'room_type_Entire','room_type_Private','room_type_Hotel']

train, test = train_test_split(df_ml, test_size = 0.3,random_state=1111)
x_train = train[feature_names]
y_train = train['price_$']
x_test = test[feature_names]
y_test = test['price_$']

from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(x_train,y_train)
predictions = model.predict(x_test)
actuals = np.array(y_test)
# linear_acc = model.score(x_test,actuals)
# print("Accuracy Score is ",linear_acc)
```

```python
from sklearn import metrics
# function to calculate adjusted R-squared
def adj_R2(r2,n,p):
    return 1-(1-r2)*(n-1)/(n-p-1)

MSE = np.sqrt(metrics.mean_squared_error(y_test,predictions))
r_train = model.score(x_train,y_train)
adj_r_train = adj_R2(model.score(x_train,y_train), train.shape[0], len(feature_names))
r_test = model.score(x_test,y_test)
adj_r_test = adj_R2(model.score(x_test,y_test), test.shape[0], len(feature_names))

print("RMSE:",MSE)
#print('mse',metrics.mean_squared_error(y_test,predictions))
print("R_squared training:",r_train)
print("Adjusted R_squared training:",adj_r_train)
print("R_squared test:",r_test)
print("Adjusted R_squared test:",adj_r_test)
```

```python
def price_interval(price):
    interval = np.linspace(0,2000,21)
    for i in range(21 - 1):
        if price >= interval[i] and price < interval[i+1]:
            return i
        elif price >= interval[21-1]:
            return 20
        else:
            continue
```

```python
y_train_ct = y_train.apply(price_interval)
y_test_ct = y_test.apply(price_interval)
```

```python
x_ = []
y_ = []

for i in range(1,10):
    model_ = tree.DecisionTreeClassifier(max_depth = i,criterion='entropy')
    model_.fit(x_train,y_train_ct)
    
    x_.append(i)
    y_.append(model_.score(x_test,y_test_ct))
    print("Maximum depth:",i)
    #print("Training R-Square",model_.score(x_train,y_train_ct))
    print("Testing R-Square",model_.score(x_test,y_test_ct))

    y_pred_ct = model_.predict(x_test)
    mse = np.sqrt(metrics.mean_squared_error(y_test_ct, y_pred_ct))
    print("Testing RMSE", mse)
plt.plot(x_,y_)
plt.xlabel('depth')
plt.ylabel('Testing R-Square')
plt.show()
```

```python
report = {'K':[],'RMSE':[],'Training R-square':[],'Testing R-square':[]}
for n in range(5,50,5):
    knn = KNeighborsRegressor(n_neighbors=n)
    knn.fit(x_train_k,y_train_k)
    pred = knn.predict(x_test_k)
    mse = np.sqrt(metrics.mean_squared_error(y_test_k,pred))
    r2_train =  knn.score(x_train_k,y_train_k)
    r2_test = knn.score(x_test_k,y_test_k)
    report['K'].append(n)
    report['RMSE'].append(mse)
    report['Training R-square'].append(r2_train)
    report['Testing R-square'].append(r2_test)
report = pd.DataFrame.from_dict(report)
report
```
